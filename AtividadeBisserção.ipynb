```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x): return np.cos(x) - np.sqrt(x)
```

```python
opt.bisect(f, 0, 1, full_output = True, maxiter = 6, disp=False)
```

```python
def fdois(x): return 5*(np.sin(x**2)) - np.exp(x/10)
```

```python
plt.plot(fdois(np.arange(0,10, 0.1)))
```

```python
print(opt.bisect(fdois, 0, 1, xtol = 1e-5, full_output = True))
```

```python
def ftres(x): return -4 + 8*x - 5*(x**2)+(x**3)
```

```python
opt.bisect(ftres, 0.5, 3)
```

```python
def fcinco(x,r,vs): 
    vt = (300*(1.3806503*10**-23))/(1.6021*10**-19)
    Id = (vs - x)/r 
    return 10**-12*(np.exp(x/vt) - 1) - Id
```
